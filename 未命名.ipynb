```python
hero_list = eval(requests.get('https://api.opendota.com/api/heroes').text)
```

```python
#hero_list = eval(requests.get('https://api.opendota.com/api/heroes').text)
raw_id2hero = {}
for hero in hero_list:
    raw_id2hero[hero['id']] = hero
```

```python
import pickle as pkl
with open('rawid2hero.pkl', 'wb') as f:
    pkl.dump(raw_id2hero, f)
```
